# 0. Importar librer√≠as

```python
import yfinance as yf
import pyfolio as pf
from datetime import datetime as dt, timedelta as td
```

# 1. Crear fechas y bajar datos

```python
start_dt = dt(1993, 1, 1)
end_dt = dt.today()
ticker = "SPY"

df = yf.download(ticker, start=start_dt, end=end_dt)
```

# 2. Calcular retornos

```python
rets = df["Adj Close"].pct_change().dropna()
```

# 3. Tear sheet

```python
pf.create_returns_tear_sheet(rets)
```
